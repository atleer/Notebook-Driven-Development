# Project

**Exercise 1** Create three notebooks
1. **notebook_1**: This notebook accesses data from a given URL (use https://uni-bonn.sciebo.de/s/3ZbmkolHLQu1ADa/download for testing) and processes it to create a file containing only the columns: `mouse`, `feedback time`, and `response time`.
2. **notebook_2**: This notebook performs statistical analysis on the processed file generated by notebook_1.
3. **notebook_3**: This notebook generates visualizations using the processed file created by notebook_1.
